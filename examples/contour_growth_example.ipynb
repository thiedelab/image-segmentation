```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

from em_analysis.contour_analysis import calculate_growth_rate_at_contour
```

```python
video_path = "/home/erik/Projects/Sungins_Data/minus_40C.tif"
video = io.imread(video_path)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
axf = np.array(axes).ravel()

for i, ax in enumerate(axf):
    ax.imshow(video[i], cmap="gray")
    ax.set_title(f"Frame {i}")
    ax.axis("off")
```

```python
# Get the contours and the growth rates

raw_contours, growth_at_contours, growth_rate_video = calculate_growth_rate_at_contour(video, pixel_size_in_nm=1)
```

```python

fig, axes = plt.subplots(3, 3, figsize=(12, 12))
axf = np.array(axes).ravel()

vmin = np.min([growth_rate_video[i].min() for i in range(len(growth_rate_video))])
vmax = np.max([growth_rate_video[i].max() for i in range(len(growth_rate_video))])
vm = np.max([vmax, abs(vmin)])

for i, ax in enumerate(axf):
    ax.imshow(video[i], cmap="gray")
    ax.scatter(raw_contours[i][:, 1], raw_contours[i][:, 0], s=1, c=growth_at_contours[i], cmap='magma', vmin=-vm, vmax=vm)
    ax.set_title(f"Frame {i}")
    ax.axis("off")
```
